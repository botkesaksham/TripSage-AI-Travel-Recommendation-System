```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV and convert it to a graph
df = pd.read_csv('data/dist_only_matrix.csv', index_col=0)
df.fillna(0, inplace=True)

# Create a graph from the dataframe
G = nx.from_pandas_adjacency(df)

# Draw the graph
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G)  # This positions nodes more aesthetically
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=700, edge_color="gray")
plt.title("POI Distance Network")
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the POI data and priority mapping
chennai_poi_df = pd.read_csv('data/chennai-poi.csv')
priority_df = pd.read_csv('data/Chennai_priority_mapping.csv')

# Ensure POIs and similarity columns are correctly aligned
poi_names = chennai_poi_df['POIs'].tolist()

# Assuming the similarity scores are stored in the second column onward
# You can choose one column based on your needs, here I'll take the first similarity column
similarities = priority_df.iloc[:, 1].tolist()  # Adjust column index if necessary

# Create the bar plot
fig = px.bar(x=poi_names, y=similarities, labels={'x': 'POI', 'y': 'Cosine Similarity'})
fig.update_layout(title="POI Cosine Similarity with User Preferences", xaxis_title="POIs", yaxis_title="Cosine Similarity")
fig.show()
```
